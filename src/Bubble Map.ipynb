````python
# map function

def geomap(df): 
    new_map = Map(location=[-14.235004, -51.92528], zoom_start=4, tiles="CartoDB dark_matter")

    colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred','lightred', 'beige', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'white', 'pink', 'lightblue', 'lightgreen']
    for x in range(df.index[0], df.index[-1]):
        # adds circles
        subset = df[df.index == x]
        for i in range(0, len(subset)):
            folium.Circle(
                          location=[subset.iloc[i]['Latitude'], subset.iloc[i]['Longitude']], 
                          popup=subset.iloc[i]['Deforestation'], 
                          radius=float(subset.iloc[i]['Deforestation'])*40,
                          color=colors[x-2004],
            ).add_to(new_map)

    return new_map
````

````python
# color key
dict(zip([x for x in range(2004, 2020)], colors))
````

---

## All data graphed

````python
geomap(df)
````

---
## year $\leq$ 2012
Focusing on deforestation up until 2012 when most deforestation rates declined:

````python
df_1 = df[df.index < 2012]
geomap(df_1)
````

---
## 2012 < year $\leq$ 2019

````python
df_2 = df[df.index > 2012]
geomap(df_2)
````

---

So far, we can see that:
* the largest amounts of deforestation took place *before* 2012
* the largest amount of reduction in deforestation took place *before* 2012
* the rates of deforestation after 2012 are pretty small (relatively) and consistent
